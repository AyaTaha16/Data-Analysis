```python
import matplotlib.pyplot as plt
import numpy as np
```

# **1. Overlapping Line Plots with Annotations**

```python
# Define the range and functions
x = np.linspace(0, 2 * np.pi, 100)
sine = np.sin(x)
cosine = np.cos(x)
tangent = np.tan(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, sine, label='Sine', color='blue', linestyle='-')
plt.plot(x, cosine, label='Cosine', color='orange', linestyle='--')
plt.plot(x, tangent, label='Tangent', color='green', linestyle=':')

# Annotate key points
plt.annotate('Cross at Ï€/4', xy=(np.pi/4, np.sqrt(2)/2),
             xytext=(np.pi/4, 0.5), arrowprops=dict(facecolor='black', shrink=0.05))
# Add grid and legend
plt.grid(color='Green', lw=0.5, ls='--')
plt.legend()
plt.title('Trigonometric Functions')
plt.xlim(0, 2 * np.pi)
plt.ylim(-3, 3)
plt.xlabel('x (radians)')
plt.ylabel('Function value')
plt.show()
```

# **2. Advanced Subplots with Different Scales**

```python
# Data for the functions
x = np.linspace(0, 10, 100)
linear = 2 * x + 1
exponential = np.exp(x)
logarithmic = np.log(x + 1)

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Linear plot
axs[0].plot(x, linear, label='Linear: 2x + 1', color='blue')
axs[0].set_title('Linear Function')
axs[0].grid()
axs[0].legend()

# Exponential plot
axs[1].plot(x, exponential, label='Exponential: e^x', color='orange')
axs[1].set_title('Exponential Function')
axs[1].set_yscale('log')
axs[1].grid()
axs[1].legend()

# Logarithmic plot
axs[2].plot(x, logarithmic, label='Logarithmic: log(x+1)', color='green')
axs[2].set_title('Logarithmic Function')
axs[2].set_yscale('symlog')
axs[2].grid()
axs[2].legend()

plt.tight_layout()
plt.show()
```

# **3.Complex Bar Chart with Custom Legends and Colors**

```python
# Data for the bar chart
students = ['Alice', 'Bob', 'Charlie', 'David']
subjects = ['Math', 'Science', 'History', 'Art']
scores = np.array([[88, 92, 85, 95],
                   [78, 85, 80, 79],
                   [95, 91, 88, 90],
                   [70, 72, 75, 74]])

# Create a horizontal grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Define bar positions
y = np.arange(len(students)) # Changed from len(subjects)

# Create bars for each student
bar_width = 0.15
for i, student in enumerate(students):
    ax.barh(y + i * bar_width, scores[i, :], bar_width, label=student) # Changed scores[:, i] to scores[i, :]

ax.set_yticks(y + bar_width * (len(students) - 1) / 2)
ax.set_yticklabels(students) # Changed from subjects
ax.set_xlabel('Scores')
ax.set_title('Test Scores of Students')
ax.legend(title='Students')
ax.grid(axis='x')

# Display the scores on the bars
for i in range(len(students)):
    for j in range(len(subjects)):
        ax.text(scores[i, j] + 1, i + j * bar_width, str(scores[i, j]), va='center')

plt.show()
```
