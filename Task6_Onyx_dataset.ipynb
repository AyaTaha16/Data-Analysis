```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/onyx.xlsx')
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
unique_store_ids = df['StoreID'].nunique()
print(f"Number of unique StoreIDs: {unique_store_ids}")
```

```python
# Calculate the IQR for StoreID (assuming it's numerical)
Q1 = df['StoreID'].quantile(0.25)
Q3 = df['StoreID'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Fill missing values with the median StoreID within the IQR range
median_store_id = df['StoreID'].median()
df['StoreID'].fillna(median_store_id, inplace=True)

print("Missing values filled with median within IQR range.")
```

```python
# Calculate the IQR for DeliveryTime
Q1_delivery = df['DeliveryTime'].quantile(0.25)
Q3_delivery = df['DeliveryTime'].quantile(0.75)
IQR_delivery = Q3_delivery - Q1_delivery

# Define the lower and upper bounds for outliers
lower_bound_delivery = Q1_delivery - 1.5 * IQR_delivery
upper_bound_delivery = Q3_delivery + 1.5 * IQR_delivery

# Replace outliers with the median within the IQR range
median_delivery_time = df['DeliveryTime'].median()
df['DeliveryTime'] = df['DeliveryTime'].apply(lambda x: median_delivery_time if x < lower_bound_delivery or x > upper_bound_delivery else x)

# Fill missing values in DeliveryTime with the range
min_delivery_time = df['DeliveryTime'].min()
max_delivery_time = df['DeliveryTime'].max()
df['DeliveryTime'].fillna(pd.Series(np.random.randint(min_delivery_time, max_delivery_time + 1, size=len(df['DeliveryTime']))), inplace=True)

print("Outliers in DeliveryTime replaced with the median within the IQR range.")
print("Missing values in DeliveryTime filled with the range.")
```

```python
# Fill missing values in ShippingMethod with the mode
shipping_method_mode = df['ShippingMethod'].mode()[0]
df['ShippingMethod'].fillna(shipping_method_mode, inplace=True)

# Fill missing values in Event with the mode
event_mode = df['Event'].mode()[0]
df['Event'].fillna(event_mode, inplace=True)

print("Missing values in ShippingMethod and Event filled with their respective modes.")

df.isnull().sum()
```

```python
# Visualize numerical features using histograms and box plots
numerical_features = df.select_dtypes(include=['number'])
for column in numerical_features.columns:
  if column not in ['TransactionID', 'CustomerID', 'StoreID', 'ProductID']:
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram of {column}')

    plt.tight_layout()
    plt.show()
```

```python
# Visualize categorical features using bar plots and count plots
categorical_features = df.select_dtypes(include=['object'])
for column in categorical_features.columns:
  plt.figure(figsize=(8, 4))
  sns.countplot(x=df[column])
  plt.title(f'Countplot of {column}')
  plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels if needed
  plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numerical features for correlation analysis
numerical_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(12, 10))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
